```python
from statsmodels.tsa.arima.model import ARIMA

# Load the dataset
df = pd.read_csv('user1_data.csv')

# Convert the ttime column to a datetime object
df['ttime'] = pd.to_datetime(df['ttime'])

# Set the ttime column as the index
df.set_index('ttime', inplace=True)

# Extract the sm column
sm = df['sm']

# Train an ARIMA model on the historical data
model = ARIMA(sm, order=(1, 1, 1))
model_fit = model.fit()

# Make predictions for the month of March
# pred = model_fit.predict(start='2023-03-01', end='2023-03-31 23:59:59', typ='levels')
pred = model_fit.predict(start='2023-03-01', end='2023-03-10 08:45:06' , freq='30T', typ='levels')

# Print the predicted values for the month of March
# print(len(pred))

# Load the actual values for March
actual = sm.loc['2023-03-01':'2023-03-31']

# Calculate the mean squared error between the actual and predicted values
mse = np.mean((actual - pred)**2)
mape = np.mean(np.abs((actual - pred) / actual)) * 100

# Calculate the root mean squared error
rmse = np.sqrt(mse)

print(f"RMSE: {rmse}")
print("MAPE:", mape*100, "%")
```
